```python
stock_prices = {
    "AAPL": 180,
    "TSLA": 250,
    "GOOG": 140,
    "MSFT": 320,
}
def calculate_investment(stock, quantity):
    if stock in stock_prices:
        return stock_prices[stock] * quantity
    else:
        return None

def stock_portfolio_tracker():
    print("Stock Portfolio Tracker")
    print("Available stocks:", list(stock_prices.keys()))

    total_value = 0
    results = []

    while True:
        stock = input("\nEnter stock symbol(or 'DONE give ' to finish): ").upper()
        if stock == "DONE":
            break

        qty = int(input("Enter quantity:"))
        investment = calculate_investment(stock, qty)
        if investment is None:
            print("stock not found!")
        else:
            print(f"{stock} -> value = ${investment}")
            total_value += investment
            results.append(f"{stock}: {qty} shares = ${investment}")
        print("\nTotal portfolio value:$" , total_value)

        save = input("save result to portfolio.txt? (yes/no): ").lower()
        if save == "yes":
            with open("portfolio.txt", "w") as file:
                for line in results:
                    file.write(line+"\n")
                    file.write(f"\nTotal Portfolio Value: ${total_value}")
            print("Saved to portfolio.txt")

stock_portfolio_tracker()
                     
        

```
